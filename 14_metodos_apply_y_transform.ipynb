````python
import pandas as pd
import numpy as np
````

## El método ```pd.DataFrame.apply()```.

Aplica una función a lo largo de un eje del dataframe

```
<dataframe>.apply(<funcion>, axis=<eje>)
```

  - `<funcion>` Una función acorde a aplicar a cada fila o columna. Es posible aplicar funciones de *NumPy*
  - `<eje>` indica el eje en el que se aplicará la *función*
    - `0` aplicará a cada columna, ***Valor por defecto***
    - `1` aplicará a cada fila

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html

**Ejemplo**

````python
# Crear un dataframe que consiste en un censo de animales
poblacion = pd.DataFrame({'Brasil':   (25, 45, 23,  67, 14),
                          'Venezuela':(31,  0, 23,   3, 34),
                          'Colombia': ( 0,  4,  3,   1,  1),
                          'Bolivia':  ( 2,  0, 12,  23, 11)},
                         index=('lobo', 'jaguar', 'coyote', 'lechuza', 'puma'))
poblacion
````

````python
# Promedio de animales por pais
poblacion.apply(lambda columna: sum(columna)/len(columna))
````

````python
# Sumar el total de cada animal
poblacion.apply(lambda fila: sum(fila), axis=1)
````

````python
# Usando funciones Numpy, promedio de animales por pais
poblacion.apply(np.mean)
````

````python
# Usando funciones propias, sumar el total de animales
def suma_lista(l):
  r = 0
  for i in l: r += i
  return r

poblacion.apply(suma_lista, axis=1)
````

+ Es compatible con broadcasting `[1,2,3] + 1`

````python
# Sumar +1 a cada elemento
poblacion.apply(lambda columna: columna + 1) #columna + 1, aplica broadcasting
````
